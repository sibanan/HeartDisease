```python
from sklearn.preprocessing import OrdinalEncoder,OneHotEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer,make_column_transformer
from sklearn.pipeline import Pipeline,make_pipeline
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB,GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.ensemble import GradientBoostingClassifier
```

```python
X_train.columns
```

```python
X_train['RestingECG'].value_counts()
```

```python
transformer = ColumnTransformer(transformers=[
    ('tnf1',StandardScaler(),['Age','RestingBP','Cholesterol','MaxHR','Oldpeak']),
    ('tnf2',OrdinalEncoder(categories=[['ASY','NAP','ATA','TA']]),['ChestPainType']),
    ('tnf3',OrdinalEncoder(categories=[['LVH','ST','Normal']]),['RestingECG']),
    ('tnf4',OneHotEncoder(sparse=False,drop='first'),['Sex','ExerciseAngina','ST_Slope'])
],remainder='passthrough')
```

```python
pipe = Pipeline([
    ('transformer',transformer)
])
```

```python
# Display Pipeline

from sklearn import set_config
set_config(display='diagram')
```

```python
pipe.fit(X_train,y_train)
```

```python
from sklearn.metrics import classification_report,plot_confusion_matrix,plot_roc_curve
```

```python
log=LogisticRegression()
svc = SVC(kernel='rbf',C=10000)
Mul = GaussianNB()
knc = KNeighborsClassifier(n_neighbors=3)
dtc = DecisionTreeClassifier()
rfc = RandomForestClassifier(n_estimators=50, random_state=2)
Gbc = GradientBoostingClassifier(n_estimators=50,random_state=2)
```

```python
clfs = {
    'LogisticRegression' :log,
    'GaussianNB':Mul,
    'SVC' : svc,
    'KNeighborsClassifier' : knc, 
    'DecisionTreeClassifier': dtc,  
    'RandomForestClassifier': rfc,  
    'GradientBoostingClassifier':Gbc,
    
}
```

```python
from sklearn.metrics import confusion_matrix,classification_report,plot_roc_curve
```

```python
from sklearn.metrics import accuracy_score,recall_score
```

```python
def train_classifier(clf,X_train,y_train,X_test,y_test):
    pipe = Pipeline([('step1',transformer),('step2',clf)])
    pipe.fit(X_train,y_train)
    y_pred = pipe.predict(X_test)
    accuracy = accuracy_score(y_test,y_pred)
    recall=recall_score(y_test,y_pred)
    con_met=confusion_matrix(y_test,y_pred)
    report=classification_report(y_test,y_pred)
    #roc=plot_roc_curve(clf,y_test,y_pred)
    
    return accuracy,recall,con_met,report#,roc
```

```python
train_classifier(svc,X_train,y_train,X_test,y_test)
```

```python
accuracy_scores = []
recall_scores = []
con_met  = []
report = []

for name,clf in clfs.items():
    
    curent_accuracy,curent_recall,curent_con_met,curent_report = train_classifier(clf,
                                                                X_train,y_train,X_test,y_test)
    
    print("For-",name)
    print("1. Accuracy----------------",curent_accuracy)
    print("2. Recall------------------",curent_recall)
    print("3. confusion Metrix--------\n"
          ,curent_con_met)
    print('4. classification_report---\n'
          ,curent_report)
    print("*"*80)
    
    accuracy_scores.append(curent_accuracy)
    recall_scores.append(curent_recall)
    con_met.append(curent_con_met)
    report.append(curent_report)
```

```python
step1 = ColumnTransformer(transformers=[
    ('tnf1',StandardScaler(),['Age','RestingBP','Cholesterol','MaxHR','Oldpeak']),
    ('tnf2',OrdinalEncoder(categories=[['ASY','NAP','ATA','TA']]),['ChestPainType']),
    ('tnf3',OrdinalEncoder(categories=[['LVH','ST','Normal']]),['RestingECG']),
    ('tnf4',OneHotEncoder(sparse=False,drop='first'),['Sex','ExerciseAngina','ST_Slope'])
],remainder='passthrough')

step2 = LogisticRegression()

pipe = Pipeline([
    ('step1',step1),
    ('step2',step2)
])

pipe=pipe.fit(X_train,y_train)

y_pred = pipe.predict(X_train)
print(pipe.score(X_train,y_train))
print(confusion_matrix(y_pred,y_train))
print(classification_report(y_pred,y_train))
#print(plot_roc_curve(pipe,y_pred,y_train))
```

```python
from sklearn.model_selection import cross_val_score
cross_val_score(pipe, X_train, y_train, cv=10, scoring='accuracy').mean()
```
